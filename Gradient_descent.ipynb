```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(100,1)
err = np .random.rand()
y = 7*x + err
print(err)
```

```python
w = 0.0
b = 0.0
lr = 0.01
```

```python
def grad_desc(x, y, w, b, lr):
    dlwdw = 0.0
    dlwdb = 0.0
    N = x.shape[0]
    for yi, xi in zip(y, x):
        dlwdw += (-2) * (yi - (w * xi + b)) * xi
        dlwdb += (-2) * (yi - (w * xi + b))
    w -= (lr * dlwdw) / N
    b -= (lr * dlwdb) / N
    return w, b
```

```python
losses = []
ws = []
bs = []
for epoch in range(400):
    w,b = grad_desc(x,y,w,b,lr)
    yhat = w*x + b
    loss = np.sum((y - yhat) ** 2) / len(y)
    losses.append(loss)
    ws.append(w.copy())
    bs.append(b.copy())
print(loss,w,b)
```

```python
plt.scatter(x, y, label='Data points')
plt.plot(x, yhat, color='red', label='Fitted Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitted Line using Gradient Descent')
plt.legend()
plt.show()

plt.plot(range(400), losses)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()

plt.plot(range(400), ws)
plt.xlabel('Epoch')
plt.ylabel('Weight')
plt.title('Weight Curve')
plt.show()


plt.plot(range(400), bs)
plt.xlabel('Epoch')
plt.ylabel('Bias')
plt.title('Bais Curve')
plt.show()
```
